```python
import os
import random
from IPython.display import display, Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Укажите путь к папке с изображениями
image_folder = r'D:\UAV_obj_det\datasets\rotate\images'

# Получаем список всех изображений
all_images = [f for f in os.listdir(image_folder) 
             if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

# Выбираем 3 случайных изображения
random_images = random.sample(all_images, min(3, len(all_images)))

# Выводим с помощью matplotlib (красиво с подписями)
plt.figure(figsize=(15, 5))

for i, img_file in enumerate(random_images, 1):
    img_path = os.path.join(image_folder, img_file)
    
    plt.subplot(1, 3, i)
    img = mpimg.imread(img_path)
    plt.imshow(img)
    plt.title(img_file)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
